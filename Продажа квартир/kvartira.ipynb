### Приводим названия столбцов в порядок

```python
#для переименования столбцов вызывают метод датафреймов rename()
df = df.rename(columns={'cityCenters_nearest': 'city_сenters_nearest', 'parks_around3000': 'parks_around_3000', 'ponds_around3000': 'ponds_around_3000'}) 
```

### Обработка пропусков

```python
# функция, получающая на вход название признака, и возвращающая число пропущенных значений
def print_nan(column):
    return print('Пропущено {:.1%}'. format(df[column].isna().value_counts()[1] / len(df), 2) + ' значений') 
```

Скорее всего цена указана в тысячах рублей,переводим в рубли.

```python
df.loc[df['last_price']<100000,'last_price'] = df.loc[df['last_price']<100000,'last_price']*1000
```

```python
# функция, получающая на вход название признака, и возвращающая границы "усов"
def det(column):
    x = df[column].describe()
    iqr = x[6] - x[4]  # межквартирльный размах  
    left = round(x[4] - 1.5 * iqr, 2) # левый и правый "ус" 
    right = round(x[6] + 1.5 * iqr, 2)
    if left < x[3]: left = x[3]   # левый "ус" не должен быть 
    if right > x[7]: right = x[7] # правый "ус" не должен быть больше максимального значения    
    return [left, right]
```

```python
# зададим функцию, вычисляющую долю аномальных значений
def anomalies(column):
    return 'Аномальные значения составляют {:.1%}'. format(len(df.loc[df[column] > det(column)[1]]) / len(df)) + ' от всей выборки' 
 
```

```python
#добавим столбец с признаком, описывающим аномально большую площадь недвижимости
df['too_total_area'] = df['total_area'] > det('total_area')[1] 
```

```python
anomalies('total_area')
```

```python
anomalies('last_price')
```

```python
#добавим столбец с признаком, описывающим аномально дорогую недвижимость
df['too_price'] = df['last_price'] > det('last_price')[1] 
```

```python
anomalies('rooms')
```

```python
#добавим столбец с признаком, описывающим недвижимость с аномально большим числом комнат
df['too_rooms'] = df['rooms'] > det('rooms')[1] 
```

```python
anomalies('ceiling_height')
```

```python
df['too_ceiling'] = (df['ceiling_height'] > det('ceiling_height')[1]) | (df['ceiling_height'] < det('ceiling_height')[0])
```

```python
#уберем редкие и выбивающиеся значения
data = df[(df['too_total_area'] == False) & (df['too_price'] == False) & (df['too_rooms'] == False) & (df['too_ceiling'] == False) ].reset_index(drop=True) 
('Доля чистых значений составила: {:.1%}'. format(len(data) / len(df)))
```
